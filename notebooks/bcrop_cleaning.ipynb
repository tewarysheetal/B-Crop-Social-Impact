```python
import pandas as pd
df = pd.read_csv("D:\\Projects\\B-Crop-Social-Impact\\data\\raw\\B-Corp-Impact-Data.csv")
```

```python
cols_to_keep = [
    "company_id",
    "company_name",
    "current_status",
    "overall_score",
    "date_first_certified",
    "date_certified",
    "industry",
    "industry_category",
    "products_and_services",
    "country",
    "state",
    "city",
    "sector",
    "size",
    "impact_area_community",
    "impact_area_customers",
    "impact_area_environment",
    "impact_area_governance",
    "impact_area_workers"
]
```

```python
df_clean = df[cols_to_keep].copy()
df_certified = df_clean[df_clean["current_status"].str.lower() == "certified"].copy()
```

```python
df_clean.to_csv("D:\\Projects\\B-Crop-Social-Impact\\data\\processed\\bcorp_clean.csv", index=False)
df_certified.to_csv("D:\\Projects\\B-Crop-Social-Impact\\data\\processed\\bcorp_certified.csv", index=False)
```
